```python
print "Top Community '8' Analysis"
print '\n'

#Create induced subgraph for Top Community "8"
G=nx.erdos_renyi_graph(100, 0.01)
G.remove_node(9)
G.remove_node(21)
G.remove_node(26)
G.remove_node(42)
G.remove_node(54)
G.remove_node(66)
G.remove_node(71)
G.remove_node(81)
G.remove_node(85)
G.add_path([9,21,26,42,54,66,71,81,85])
subgraph_1 = G.subgraph([9,21,26,42,54,66,71,81])
print "Subgraph & Hierarchy Plot:"
print subgraph_1.edges()
nx.draw(subgraph_1)
plt.show()
print '\n'

print "Subgraph 1 Info:"
print nx.info(subgraph_1)
print '\n'

print "Number of Cliques in Subgraph 1:"
print nx.number_of_cliques(subgraph_1)
print '\n'

print "K-Cliques in Supgraph 1:"
print [l for l in nx.k_clique_communities(subgraph_1,2)]
print '\n'

print "Max Cliques in Subgraph 1:"
print [l for l in nx.find_cliques(subgraph_1)]
print '\n'

print "Density:"
print nx.density (subgraph_1)
print '\n'

print "Degree Centrality:"
dc = nx.degree_centrality(subgraph_1)
import operator
sorted_dc = sorted(dc.items(), key=operator.itemgetter(1),reverse=True)
print sorted_dc
print '\n'

print "Betweeness Centrality:"
dc = nx.betweenness_centrality(subgraph_1)
import operator
sorted_dc = sorted(dc.items(), key=operator.itemgetter(1),reverse=True)
print sorted_dc
print '\n'

print "Closeness Centrality:"
dc = nx.closeness_centrality(subgraph_1)
import operator
sorted_dc = sorted(dc.items(), key=operator.itemgetter(1),reverse=True)
print sorted_dc
print '\n'

print "Average Shortest Path:"
print nx.average_shortest_path_length(subgraph_1)
print '\n'

print "Average Clustering Coefficient:"
print nx.average_clustering(subgraph_1)
print '\n'

print "Modularity Analysis of Community 8:"
part = community.best_partition(G)
print(community.modularity(part, G))
print '\n'

print "Hierarchy of Top Community 8:"
dendrogram = community.generate_dendrogram(subgraph_1)
for level in range(len(dendrogram)) :
    print("Partition at level", level, "is", community.partition_at_level(dendrogram, level))   
                
print '\n'
nx.draw(subgraph_1, pos=nx.spring_layout(subgraph_1), node_size = 20)
plt.show()
```

```python
#better with karate_graph() as defined in networkx example.
#erdos renyi don't have true community structure
#first compute the best partition
partition = community.best_partition(G)

#drawing
size = float(len(set(partition.values())))
pos = nx.spring_layout(G)
count = 0.
for com in set(partition.values()) :
    count = count + 1.
    list_nodes = [nodes for nodes in partition.keys()
                                if partition[nodes] == com]
    nx.draw_networkx_nodes(G, pos, list_nodes, node_size = 20,
                                node_color = str(count / size))


nx.draw_networkx_edges(G,pos, alpha=0.5)
plt.show()
```
